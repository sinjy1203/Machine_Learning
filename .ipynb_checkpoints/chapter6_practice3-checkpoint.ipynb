# 연습문제 3번

## SVM  Support Vector Machine

## linear kernel, gauss kernel

### load iris data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()

x = iris['data'][:, [0,1]] 
y = iris['target']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### linear kernel

```python
from sklearn import svm

linear_svc = svm.SVC(kernel='linear')
linear_svc.fit(x_train, y_train)

print('학습데이터 정확도 = %.2f' % linear_svc.score(x_train, y_train))
print('시험데이터 정확도 = %.2f' % linear_svc.score(x_test, y_test))
```

```python
# 시각화
import mglearn

plt.figure(figsize=(8, 5))
mglearn.plots.plot_2d_classification(linear_svc, x_train, alpha=0.1)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

### gauss kernel  ( rbf )

```python
from sklearn import svm

rbf_svc = svm.SVC(kernel='rbf')
rbf_svc.fit(x_train, y_train)

print('학습데이터 정확도 = %.2f' % rbf_svc.score(x_train, y_train))
print('시험데이터 정확도 = %.2f' % rbf_svc.score(x_test, y_test))
```

```python
# 시각화
import mglearn

plt.figure(figsize=(8, 5))
mglearn.plots.plot_2d_classification(rbf_svc, x_train, alpha=0.1)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```
