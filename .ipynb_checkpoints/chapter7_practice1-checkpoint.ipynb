# 연습문제 1번

## MLE  Maximum Likelihood Estimation

### 3.0 data load

```python
import sklearn
import numpy as np
import pandas as pd
from pathlib import Path

data_dir = Path('./data/chapter4_dataset.csv')
df = pd.read_csv(data_dir)
df = df.drop(['Idx'], axis=1)
```

```python
df
```

```python
data = df[['color', 'root', 'knocks']]
label = df['label']
```

### class-conditional probability 
$P(x_i | c) = \frac{|D_{c, x_i}|}{|D_c|}$ 
### from i.i.d

```python
columns = data.columns
columns_att = {column : data[column].unique() for column in columns}

p_class_conditional = {}
for column in columns:
    att_dict = {}
    for att in columns_att[column]:
        is_att = data[column] == att
        
        label_dict = {}
        for i in range(2):
            is_label = label == i
            is_att_label = is_att & is_label
            label_dict[i] = len(data[is_att_label]) / len(label[is_label])
        
        att_dict[att] = label_dict
    p_class_conditional[column] = att_dict
```
