## AODE  Averaged One-Denpendent Estimator

### 3.0 data load

```python
import sklearn
import numpy as np
import pandas as pd
from pathlib import Path

data_dir = Path('./data/chapter4_dataset.csv')
df = pd.read_csv(data_dir)
df = df.drop(['Idx'], axis=1)
df
```

```python
train_x = df.drop(['density', 'sugar_ratio', 'label'], axis=1)
train_y = df['label']

test = df.iloc[:1]
```

```python
test
```

$P(c|x) ∝ \sum_{i=1}^d P(c, x_i) \prod_{j=1}^d P(x_j|c, x_i)$

$P(c, x_i) = \frac{|D_{c, x_i}| + 1}{|D| + N * N_i}$

$P(x_j|c, x_i) = \frac{D_{c,x_i,x_j} + 1}{D_{c, x_i} + N_j}$

```python
p_c_x = {}
D_label = {0: train_y == 0, 1: train_y == 1}
D = len(train_x)

for label in range(2):
    p_sum = 0
    for att_i in train_x.columns:
        D_xi = train_x[att_i] == test[att_i][0]
        D_cxi = D_label[label] & D_xi
        n_i = len(train_x[att_i].unique())
        p_ = (len(train_x[D_cxi]) + 1) / (D + 2 * n_i)
        
        for att_j in train_x.columns:
            D_xj = train_x[att_j] == test[att_j][0]
            D_cxixj = D_cxi & D_xj
            n_j = len(train_x[att_j].unique())
            p_ *= (len(train_x[D_cxixj]) + 1) / (len(train_x[D_cxi]) + n_j)
        p_sum += p_
    p_c_x[label] = p_sum

p_c_x
```
